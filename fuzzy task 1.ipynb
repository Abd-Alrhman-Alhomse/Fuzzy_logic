```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
```

```python
x = np.arange(0, 10.01, 0.01) 


a, b, c, d = 2, 3, 7, 8  


mf = fuzz.trapmf(x, [a, b, c, d])


plt.plot(x, mf, label='Trapezoidal MF')


plt.xlabel('Input Value')
plt.ylabel('Degree of Membership')
plt.title('Trapezoidal Membership Function')


plt.grid(True)
plt.legend()
plt.xlim([0, 10])  
plt.ylim([0, 1.1])   

plt.show()
```

2. Write a Matlab script to solve the next exercise:
Consider "Speed" as a linguistic variable whose universe of discourse is [0,140] km/h. Define five linguistic values and plot the corresponding fuzzy sets on the same axis (use "hold on"). The fuzzy sets must form a fuzzy partition. Print in the command line the membership degrees to each fuzzy set, for the following values of the speed: 10 km/h, 52 km/h, 85 km/h and 100 km/h.

```python
speed = np.arange(0, 140.01, 0.01) 


low = fuzz.trimf(speed, [0, 0, 30])
medium_low = fuzz.trimf(speed, [20, 40, 60])
medium = fuzz.trimf(speed, [50, 70, 90])
medium_high = fuzz.trimf(speed, [80, 100, 120])
high = fuzz.trimf(speed, [110, 140, 140])




speed_values = [10, 52, 85, 100]
for value in speed_values:
  membership_low = float(low[int(value * 100)])
  membership_medium_low = float(medium_low[int(value * 100)])
  membership_medium = float(medium[int(value * 100)])
  membership_medium_high = float(medium_high[int(value * 100)])
  membership_high = float(high[int(value * 100)])


  print(f"Speed: {value} km/h - Membership degrees:")
  print(f"\tLow: {membership_low}")
  print(f"\tMedium Low: {membership_medium_low}")
  print(f"\tMedium: {membership_medium}")
  print(f"\tMedium High: {membership_medium_high}")
  print(f"\tHigh: {membership_high}")

 
    
plt.plot(speed, low, label='Low')
plt.plot(speed, medium_low, label='Medium Low')
plt.plot(speed, medium, label='Medium')
plt.plot(speed, medium_high, label='Medium High')
plt.plot(speed, high, label='High')




plt.xlabel('Speed (km/h)')
plt.ylabel('Degree of Membership')
plt.title('Fuzzy Sets for Speed')
plt.legend()
plt.grid(True)
plt.xlim([0, 140])

plt.show()
```

1. Write a Matlab script to plot the result of the intersection of fuzzy sets A and B, using the "min" operator. Using the "Data Cursor" in the figure toolbar, mark on the graph the membership degrees for the next input values: x=4,5,6,7.

```python
x = np.arange(0, 10.01, 0.1)  


A = fuzz.gaussmf(x,  4 , 1)


B = fuzz.trimf(x, [3, 6.5, 9])  


plt.plot(x, A, label='Gaussian MF (A)')
plt.plot(x, B, label='Triangular MF (B)')
plt.xlabel('x')
plt.ylabel('Degree of Membership')
plt.title('Membership Functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
intersection = np.minimum(A, B)



plt.plot(x, intersection, label='Intersection (min)')
plt.xlabel('x')
plt.ylabel('Degree of Membership')
plt.title('Intersection Function' )
plt.legend()
plt.grid(True)



plt.show()
```

Consider "Speed" as a linguistic variable whose universe of discourse is [0,140]km/h. Define five linguistic values and create the corresponding fuzzy sets. Use the 'zmf', 'gbellmf' and 'smf' membership function types.

```python

speed = np.arange(0, 140.01, 0.01) 


low = fuzz.zmf(speed, 20, 40)  
medium_low = fuzz.gbellmf(speed,10, 2, 40) 
medium = fuzz.gbellmf(speed, 20 , 2 , 70)  
medium_high = fuzz.gbellmf(speed, 10, 2, 100) 
high = fuzz.smf(speed, 100 , 120)




plt.plot(speed, low, label='Low (zmf)')
plt.plot(speed, medium_low, label='Medium Low (gbellmf)')
plt.plot(speed, medium, label='Medium (gbellmf)')
plt.plot(speed, medium_high, label='Medium High (gbellmf)')
plt.plot(speed, high, label='High (smf)')
plt.xlabel('Speed (km/h)')
plt.ylabel('Degree of Membership')
plt.title('Fuzzy Sets for Speed (Different Membership Functions)')
plt.legend()
plt.grid(True)
plt.xlim([0, 140])
plt.show()
```
